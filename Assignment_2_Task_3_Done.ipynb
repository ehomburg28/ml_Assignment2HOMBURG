## Exercise 1: Probability to Log(Odds)

Complete the function to convert probability to log(odds).

Hint: odds = p / (1 - p)

```python
def prob_to_log_odds(p):
    # Step 1: Calculate odds
    odds = p / (1 - p)

    # Step 2: Calculate log of odds
    log_odds = np.log(odds)

    return log_odds

# Test your function
print("p = 0.5 -> log(odds) =", prob_to_log_odds(0.5))   # Expected: 0.0
print("p = 0.8 -> log(odds) =", prob_to_log_odds(0.8))   # Expected: ~1.386
```

## Exercise 2: Log(Odds) to Probability

Complete the function to convert log(odds) back to probability.

Hint: p = 1 / (1 + e^(-log_odds))

```python
def log_odds_to_prob(log_odds):
    # Use the sigmoid formula
    p = 1 / (1 + np.exp(-log_odds))

    return p

# Test your function
print("log(odds) = 0  -> p =", log_odds_to_prob(0))    # Expected: 0.5
print("log(odds) = 3  -> p =", log_odds_to_prob(3))    # Expected: ~0.95
```

## Exercise 3: Calculate Likelihood

You have 4 students. Their actual result and predicted probabilities of passing are:

| Student | Actual Result | Predicted Probability |
|---------|---------------|-----------------------|
| A | Pass (1) | 0.9 |
| B | Pass (1) | 0.6 |
| C | Fail (0) | 0.2 |
| D | Fail (0) | 0.4 |

Calculate the total likelihood.

```python
# For passing students, likelihood = predicted probability
# For failing students, likelihood = 1 - predicted probability

# Student A (Pass)
likelihood_A = 0.9

# Student B (Pass)
likelihood_B = 0.6

# Student C (Fail)
likelihood_C = 0.2

# Student D (Fail)
likelihood_D = 0.4

# Total likelihood = product of all
total_likelihood = likelihood_A * likelihood_B * likelihood_C * likelihood_D

print("Total Likelihood:", total_likelihood)   # Expected: 0.324
```

## Exercise 4: Train a Model

Use the dataset below to train a Logistic Regression model.

Features: Hours Studied, Hours Slept

Target: 0 = Fail, 1 = Pass

```python
# Dataset
X = np.array([
    [1, 4],
    [2, 5],
    [3, 5],
    [5, 7],
    [6, 7],
    [7, 8],
])

y = np.array([0, 0, 0, 1, 1, 1])

# Step 1: Create the model
model = LogisticRegression()

# Step 2: Train the model
model.fit(X, y)

print("Model trained!")
```

## Exercise 5: Make a Prediction

Use your trained model to predict for a new student:
- Hours Studied: 4
- Hours Slept: 6

```python
# Create new student data
new_student = np.array([[4, 6]])

# Get predicted probability
probability = model.predict_proba(new_student)[0][1]

# Get prediction (0 or 1)
prediction = model.predict(new_student)[0]

print("Probability of Passing:", probability)
print("Prediction:", prediction)
```
